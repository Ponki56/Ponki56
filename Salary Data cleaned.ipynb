```python
import pandas as pd
import numpy as np
df = pd.read_csv('D:\Data Science Projects\glassdoor_jobs.csv')
```

```python
df.dtypes
```

```python
df = df[df['Salary Estimate']!= '-1']
```

```python
df
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary
```

```python
minus_kd = salary.apply(lambda x:x.replace('K',' ').replace('$',' '))
```

```python
minus_kd
```

```python
df['hourly']= df['Salary Estimate'].apply(lambda x: 1 if 'per hour'in x.lower() else 0)
```

```python
df['employer provided salary'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary'in x.lower()else 0)
```

```python
pd.set_option('display.max_rows',1000)
```

```python
df['employer provided salary']
```

```python
df['hourly']
```

```python
minus_hr = minus_kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary','').replace(':',''))
```

```python
df['min_salary']= minus_hr.apply(lambda x: int(x.split('-')[0]))
```

```python
df['min_salary']
```

```python
df['max_salary']= minus_hr.apply(lambda x: int(x.split('-')[1]))
```

```python
df['avg_salary'] = (df.min_salary + df.max_salary)/2
```

```python
df['Company Name'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3], axis =1)
```

```python
df['Job State']= df['Location'].apply(lambda x: x.split(',')[1])
```

```python
df['Job State'].value_counts()
```

```python
df['same_state'] = df.apply(lambda x:1 if x.Location==x.Headquarters else 0, axis = 1)
```

```python
df['Company_age']= df['Founded'].apply(lambda x: x if x<1 else 2022-x)
```

```python
df['python_dep'] = df['Job Description'].apply(lambda x:1 if 'python'in x.lower() else 0)
```

```python
df.python_dep.value_counts()
```

```python
df['R']= df['Job Description'].apply(lambda x: 1 if 'r studio'in x.lower() else 0)
```

```python
df.R.value_counts()
```

```python
df['spark']= df['Job Description'].apply(lambda x: 1 if 'spark'in x.lower() else 0)
```

```python
df['Excel']= df['Job Description'].apply(lambda x: 1 if 'excel'in x.lower() else 0)
```

```python
df_out = df.drop(['Unnamed: 0'], axis=1) #dropping the unnamed column
```

```python
df_out.to_csv('Salary_data_cleaned', index=False)
```
